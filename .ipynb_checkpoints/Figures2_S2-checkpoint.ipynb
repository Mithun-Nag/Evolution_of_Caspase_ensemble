**FETCH DEGREE AND BETWEENESS VALUES CALCUALTED FROM RAW DATA IN Centrality_Analysis Folder**

```python
import os
import pandas as pd

# === Paths ===
base_path = "Centrality_Analysis"
output_path = "Figure_2_Pipe"
os.makedirs(output_path, exist_ok=True)
final_output_file = os.path.join(output_path, "Avg_DC_BC_all.xlsx")

# === Function to extract 'Average' columns ===
def extract_averages(condition_label):
    output_dict = {}
    types = ["Degree", "Betweeness"]
    subtypes = ["Active", "Dimeric", "Monomeric"]

    for typ in types:
        folder_name = f"{typ}_{condition_label}"
        for subtype in subtypes:
            file_path = os.path.join(base_path, condition_label, folder_name, subtype, f"{subtype}.csv")
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
                if "Average" in df.columns:
                    colname = f"Avg_{typ}_{condition_label}_{subtype}"
                    output_dict[colname] = df["Average"]
                else:
                    print(f"⚠️ 'Average' column not found in {file_path}")
            else:
                print(f"⚠️ File not found: {file_path}")
    return output_dict

# === Gather average data ===
avg_data = {}
avg_data.update(extract_averages("No_Urea"))
avg_data.update(extract_averages("Urea"))

# === Combine into one DataFrame ===
avg_df = pd.DataFrame(avg_data)

# === Append Consurf annotations ===
consurf_file = os.path.join("Consurf_grades", "Conservation_Annotation_Output.csv")
if os.path.exists(consurf_file):
    consurf_df = pd.read_csv(consurf_file)[["short_SS", "Conservation_Score", "Conservation"]]
    final_df = pd.concat([avg_df, consurf_df], axis=1)
else:
    print(f"⚠️ Consurf file not found: {consurf_file}")
    final_df = avg_df

# === Save final output ===
final_df.to_excel(final_output_file, index=False)
print(f"✅ Final output saved to: {final_output_file}")
```

**Active conformation in water violin plot for individual plots**

Note: change column name Avg_Degree_No_Urea_Active to plot monomeric and dimeric, refer the excel file in Figure_2_Pipe folder

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# === Load Data ===
df = pd.read_excel("Figure_2_Pipe/Avg_DC_BC_all.xlsx")

# === Normalize and map secondary structure labels ===
df['short_SS'] = df['short_SS'].astype(str).str.strip().str.capitalize()
category_map = {
    'B loop': 'Bottom\nLoops',
    'Bs': 'Beta\nSheets',
    'Loop': 'Top\nLoops',
    'Helix': 'Helices',
    'Shortbs': 'Short\nBeta\nSheet'
}
df['short_SS'] = df['short_SS'].map(category_map)
df = df.dropna(subset=['short_SS'])

# === Set order and color palette ===
desired_order = ['Bottom\nLoops', 'Beta\nSheets', 'Top\nLoops', 'Helices', 'Short\nBeta\nSheet']
set1_colors = sns.color_palette('Set1', n_colors=5)
custom_color_order = [set1_colors[1], set1_colors[4], set1_colors[2], set1_colors[0], set1_colors[3]]
custom_palette = dict(zip(desired_order, custom_color_order))

# === Select the plot ===
col_name = "Avg_Degree_No_Urea_Active"
title = "Active Conformation in Water"

# === Create the plot ===
sns.set_context('paper', font_scale=2.5)
sns.set_style("white")
plt.figure(figsize=(10, 8))

sub_df = df.dropna(subset=[col_name])
sns.violinplot(
    x="short_SS",
    y=col_name,
    data=sub_df,
    order=desired_order,
    palette=custom_palette,
    width=0.5
)

plt.title(title, fontsize=22)
plt.xlabel("")
plt.ylabel("Degree", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=14)
plt.gca().tick_params(axis='x', direction='out', length=12, width=2)
plt.gca().tick_params(axis='y', direction='out', length=12, width=2)

# Thicken plot frame
for spine in ['left', 'bottom', 'right', 'top']:
    plt.gca().spines[spine].set_linewidth(2)

# === Save and Show ===
plt.tight_layout()
plt.savefig("Figure_2_Pipe/Figure2_Single.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

**All plots side by side from Main Figure 2**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# === Load Data ===
df = pd.read_excel("Figure_2_Pipe/Avg_DC_BC_all.xlsx")

# === Normalize and map secondary structure labels ===
df['short_SS'] = df['short_SS'].astype(str).str.strip().str.capitalize()
category_map = {
    'B loop': 'Bottom\nLoops',
    'Bs': 'Beta\nSheets',
    'Loop': 'Top\nLoops',
    'Helix': 'Helices',
    'Shortbs': 'Short\nBeta'
}
df['short_SS'] = df['short_SS'].map(category_map)

# === Remove rows with missing structure or degree ===
df = df.dropna(subset=['short_SS'])

# === Set order and color palette ===
desired_order = ['Bottom\nLoops', 'Beta\nSheets', 'Top\nLoops', 'Helices', 'Short\nBeta']
set1_colors = sns.color_palette('Set1', n_colors=5)
custom_color_order = [set1_colors[1], set1_colors[4], set1_colors[2], set1_colors[0], set1_colors[3]]
custom_palette = dict(zip(desired_order, custom_color_order))

# === Plot Setup ===
sns.set_context('paper', font_scale=2.5)
sns.set_style("white")

fig, axes = plt.subplots(2, 3, figsize=(22, 14), sharey=True)

# === Plot map ===
plot_map = [
    ("Avg_Degree_No_Urea_Active", "Active Conformation in Water"),
    ("Avg_Degree_No_Urea_Dimeric", "Dimeric Conformation in Water"),
    ("Avg_Degree_No_Urea_Monomeric", "Monomeric Conformation in Water"),
    ("Avg_Degree_Urea_Active", "Active Conformation in Urea"),
    ("Avg_Degree_Urea_Dimeric", "Dimeric Conformation in Urea"),
    ("Avg_Degree_Urea_Monomeric", "Monomeric Conformation in Urea"),
]

# === Create the grid of plots ===
for idx, (col_name, title) in enumerate(plot_map):
    ax = axes[idx // 3, idx % 3]
    if col_name not in df.columns:
        continue

    sub_df = df.dropna(subset=[col_name])
    sns.violinplot(
        x="short_SS",
        y=col_name,
        data=sub_df,
        order=desired_order,
        palette=custom_palette,
        width=0.5,
        ax=ax
    )

    ax.set_title(title, fontsize=22)
    ax.set_xlabel("")
    ax.set_ylabel("Degree" if idx % 3 == 0 else "", fontsize=18)

    ax.set_xticks(range(len(desired_order)))
    ax.set_xticklabels(desired_order, fontsize=16)
    ax.tick_params(axis='x', which='both', bottom=True, top=False, direction='out', length=12, width=2)
    ax.tick_params(axis='y', which='both', left=True, right=False, direction='out', length=12, width=2)

    for spine in ['left', 'bottom', 'right', 'top']:
        ax.spines[spine].set_linewidth(2)

# === Save and show ===
plt.subplots_adjust(wspace=0.3, hspace=0.4, top=0.95, bottom=0.1)

plt.savefig("Figure_2_Pipe/Figure2_Grid.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

**Supplementary figure S2 with a single plot**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# === Load Data ===
df = pd.read_excel("Figure_2_Pipe/Avg_DC_BC_all.xlsx")

# === Clean secondary structure column ===
df['short_SS'] = df['short_SS'].astype(str).str.strip().str.capitalize()

# === Map secondary structure to readable form ===
category_map = {
    'B loop': 'B Loop',
    'Bs': 'BS',
    'Loop': 'Loop',
    'Helix': 'Helix',
    'Shortbs': 'ShortBS'
}
df['short_SS'] = df['short_SS'].map(category_map)

# === Drop rows with missing values ===
df = df.dropna(subset=['short_SS', 'Avg_Degree_No_Urea_Active', 'Conservation'])

# === Order and colors ===
order = ['B Loop', 'BS', 'Loop', 'Helix', 'ShortBS']
custom_labels = ["Bottom\nLoops", "Beta\nSheets", "Top\nLoops", "Helices", "Short\nBeta\nSheets"]
hue_order = ['Highly Conserved', 'Variable', 'Intermediate']
palette = {
    'Highly Conserved': '#1f77b4',
    'Variable': '#ff7f0e',
    'Intermediate': '#2ca02c'
}

# === Plot settings ===
sns.set_context('paper', font_scale=2.5)
sns.set_style("white")

plt.figure(figsize=(14, 8))
ax = sns.violinplot(
    x='short_SS',
    y='Avg_Degree_No_Urea_Active',
    hue='Conservation',
    data=df,
    order=order,
    hue_order=hue_order,
    palette=palette,
    scale="width",
    width=0.5,
    cut=2
)

# === Labels and style ===
plt.title("Active Conformation in Water", fontsize=24)
plt.xlabel("Secondary Structure", fontsize=20)
plt.ylabel("Degree", fontsize=20)
ax.set_xticklabels(custom_labels, fontsize=18)
plt.yticks(fontsize=18)
plt.legend(title="Conservation", title_fontsize=16, fontsize=14)
plt.tight_layout()

# === Save and show ===
plt.savefig("Figure_2_Pipe/Figure2_OnePanel.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# === Load Data ===
df = pd.read_excel("Figure_2_Pipe/Avg_DC_BC_all.xlsx")

# === Normalize and map secondary structure labels ===
df['short_SS'] = df['short_SS'].astype(str).str.strip().str.capitalize()
category_map = {
    'B loop': 'B Loop',
    'Bs': 'BS',
    'Loop': 'Loop',
    'Helix': 'Helix',
    'Shortbs': 'ShortBS'
}
df['short_SS'] = df['short_SS'].map(category_map)

# === Rename short_SS for x-axis display ===
label_map = {
    'B Loop': 'Bottom\nLoops',
    'BS': 'Beta\nSheets',
    'Loop': 'Top\nLoops',
    'Helix': 'Helices',
    'ShortBS': 'Short\nBeta\nSheets'
}
df['short_SS'] = df['short_SS'].map(label_map)

# === Drop rows missing essential data ===
df = df.dropna(subset=['short_SS', 'Conservation'])

# === Setup categories and palette ===
ss_order = ['Bottom\nLoops', 'Beta\nSheets', 'Top\nLoops', 'Helices', 'Short\nBeta\nSheets']
hue_order = ['Highly Conserved', 'Variable', 'Intermediate']
palette = {
    'Highly Conserved': '#1f77b4',
    'Variable': '#ff7f0e',
    'Intermediate': '#2ca02c'
}

# === Plot Setup ===
sns.set_context('paper', font_scale=2.5)
sns.set_style("white")
fig, axes = plt.subplots(2, 3, figsize=(22, 14), sharey=True)

# === Define plots ===
plot_map = [
    ("Avg_Degree_No_Urea_Active", "Active Conformation in Water"),
    ("Avg_Degree_No_Urea_Dimeric", "Dimeric Conformation in Water"),
    ("Avg_Degree_No_Urea_Monomeric", "Monomeric Conformation in Water"),
    ("Avg_Degree_Urea_Active", "Active Conformation in Urea"),
    ("Avg_Degree_Urea_Dimeric", "Dimeric Conformation in Urea"),
    ("Avg_Degree_Urea_Monomeric", "Monomeric Conformation in Urea"),
]

# === Generate the plots ===
for idx, (col_name, title) in enumerate(plot_map):
    ax = axes[idx // 3, idx % 3]
    if col_name not in df.columns:
        continue

    sub_df = df.dropna(subset=[col_name])

    sns.violinplot(
        x='short_SS',
        y=col_name,
        hue='Conservation',
        data=sub_df,
        order=ss_order,
        hue_order=hue_order,
        palette=palette,
        scale='width',
        width=0.5,
        cut=2,
        ax=ax
    )

    ax.set_title(title, fontsize=22)
    ax.set_xlabel("Secondary Structure", fontsize=18)
    ax.set_ylabel("Degree" if idx % 3 == 0 else "", fontsize=18)
    ax.set_xticklabels(ss_order, fontsize=16)
    ax.tick_params(axis='x', which='both', bottom=True, top=False, direction='out', length=12, width=2)
    ax.tick_params(axis='y', which='both', left=True, right=False, direction='out', length=12, width=2)

    for spine in ['left', 'bottom', 'right', 'top']:
        ax.spines[spine].set_linewidth(2)

    if idx % 3 != 2:
        ax.legend_.remove()

# === Adjust layout ===
plt.subplots_adjust(wspace=0.3, hspace=0.4, top=0.95, bottom=0.1)
plt.legend(title="Conservation", title_fontsize=16, fontsize=14, loc='upper center', bbox_to_anchor=(1.1, 1.05))

# === Save and show ===
plt.savefig("Figure_2_Pipe/Figure2_Grid_ColoredByConservation.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```
