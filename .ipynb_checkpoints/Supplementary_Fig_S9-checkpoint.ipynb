**Import Necessary modules**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

**Fetch requried files for S9 figure: Extract and Combine Network Subgroup and Centrality Metrics**

This notebook reads two Excel files:
1. `Network_Subgroups_fig_S9.xlsx` (which has network subgroups) from the `Supp_S9_High_DC_BC` folder
2. `Avg_DC_BC_all.xlsx` from the `Figure_2_Pipe` folder to fetch Degree and Betweeness values that are required for Pair plots 

```python
import os
# === Define directories and file paths ===
working_dir = os.getcwd()

# Input files
s9_file = os.path.join(working_dir, "Supp_S9_High_DC_BC", "Network_Subgroups_fig_S9.xlsx")
avg_file = os.path.join(working_dir, "Figure_2_Pipe", "Avg_DC_BC_all.xlsx")

# Output file
output_file = os.path.join(working_dir, "Supp_S9_High_DC_BC", "Fig_S9_DC_BC_Monomer_Dimer_no_Urea.xlsx")

# === Read the Excel files ===
df_s9 = pd.read_excel(s9_file)
df_avg = pd.read_excel(avg_file)

# === Extract columns ===
network_col = df_s9["Network_Sub_groups"]

selected_cols = df_avg[[
    "Avg_Betweeness_No_Urea_Monomeric",
    "Avg_Betweeness_No_Urea_Dimeric",
    "Avg_Degree_No_Urea_Dimeric",
    "Avg_Degree_No_Urea_Monomeric"
]]

# === Combine them ===
df_final = pd.concat([network_col, selected_cols], axis=1)

# === Write to Excel ===
df_final.to_excel(output_file, index=False)

print(f"✅ Output saved to: {output_file}")
```

**Pair Plot (Degree vs Betweeness) S9 for Dimeric conformations in water**

```python
#Figure S9 A

Avg_all = pd.read_excel('Supp_S9_High_DC_BC/Fig_S9_DC_BC_Monomer_Dimer_no_Urea.xlsx')

# Define a custom palette for the hue categories
custom_palette = {
    'B Loop': '#9ECAE1',
    'High_DC_BC': '#CB181D',
    'BS': '#BDBDBD',
    'Helix': '#BDBDBD',
    'Loop': '#C7E9C0',  # Using 'lime' for light green
    'IL': '#FDE0DD',    # Using 'khaki' for cream
    'Mutated': '#41AB5D'
}
#for conserved grey: #525252
# Set the context with increased font size
sns.set_context('paper', font_scale=3)

# Create the joint plot with custom palette
joint_plot = sns.jointplot(
    x='Avg_Betweeness_No_Urea_Dimeric', y='Avg_Degree_No_Urea_Dimeric',
    data=Avg_all, hue='Network_Sub_groups',
    height=10, legend=False, s=200,
    palette=custom_palette  # Use the custom palette
)
joint_plot.set_axis_labels('', '')  # Remove x and y axis labels

# Remove x and y axis legends
joint_plot.ax_marg_x.set_axis_off()
joint_plot.ax_marg_y.set_axis_off()

# Set x-axis and y-axis limits to only show the first quadrant
plt.xlim(-500, 6500)
plt.ylim(-2, 70)

# Additional customizations
plt.yticks([0, 10, 20, 30, 40,50,60,70])
plt.xticks([0, 2000, 4000, 6000])
plt.axhline(y=20, color='black', linestyle='--', linewidth=1)  # Horizontal line at y=20
plt.axvline(x=1000, color='black', linestyle='--', linewidth=1)  # Vertical line at x=1000
plt.tick_params(width=2, length=8)  # Set thicker and longer tick marks
# Show the plot
plt.show()
```
